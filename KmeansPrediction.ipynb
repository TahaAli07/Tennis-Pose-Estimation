```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv('export_dataframe.csv',sep=',')
print(df.head(100))
print(df.columns)
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
kmeans = KMeans(n_clusters=3) 
kmeans.fit(df)
```

```python
labels = kmeans.predict(df)
centroids = kmeans.cluster_centers_
```

```python
labels
unique_elements, counts_elements = np.unique(labels, return_counts=True)
labelfrequency=np.asarray((unique_elements, counts_elements))
```

```python
max=0
minimum=100
k=0
bestcluster=0
while k<3:
    print(labelfrequency[1][k])
    if labelfrequency[1][k]>max :
        max=labelfrequency[1][k]
        bestcluster=k
    if labelfrequency[1][k]<minimum :
        minimum=labelfrequency[1][k]
        worstcluster=k
    k=k+1
print("Best Shots: ",bestcluster)
print("Worst Shots: ",worstcluster)
```
